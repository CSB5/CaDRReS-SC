### Compare mat_norm

````python
cell_id_list = ['RHH2180', 'RHH2185', 'RHO715']
````

````python
count_df = count_df.loc[~count_df.index.duplicated(keep='first')]
````

````python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10, 3))

common_gene_list = list(set(mat_norm_df.index).intersection(count_df.index))

for ax, cell_id in zip(axes.flatten(), cell_id_list):
    
    x = mat_norm_df.loc[common_gene_list, cell_id]
    y = count_df.loc[common_gene_list, cell_id]

    sns.scatterplot(x, np.log(y+1), ax=ax, s=10, alpha=0.6)
    ax.set_xlabel('mat_norm')
    ax.set_ylabel('log(count)')
    ax.set_title(cell_id)

plt.tight_layout()
````

````python
temp_tpm_df = tpm_df.copy()
temp_tpm_df.index = [gene_id_name_dict[g] for g in temp_tpm_df.index]
temp_tpm_df = temp_tpm_df.loc[~temp_tpm_df.index.duplicated(keep='first')]
````

````python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10, 3))

common_gene_list = list(set(mat_norm_df.index).intersection(temp_tpm_df.index))

for ax, cell_id in zip(axes.flatten(), cell_id_list):
    
    x = mat_norm_df.loc[common_gene_list, cell_id]
    y = temp_tpm_df.loc[common_gene_list, cell_id]

    sns.scatterplot(x, np.log(y+1), ax=ax, s=10, alpha=0.6)
    ax.set_xlabel('mat_norm')
    ax.set_ylabel('log(TPM)')
    ax.set_title(cell_id)

plt.tight_layout()
````
